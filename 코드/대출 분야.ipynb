```python
loan = loan.loc[loan['출판사'].notnull()] #결측치 제거
loan = loan.loc[loan['ISBN'].notnull()] #결측치 제거
loan = loan.loc[loan['ISBN부가기호'].notnull()] #결측치 제거
loan = loan.loc[loan['KDC'].notnull()] #결측치 제거
loan = loan.loc[loan['성별'].notnull()] #결측치 제거
loan = loan.loc[loan['연령'].notnull()] #결측치 제거
loan = loan.loc[loan['지역'].notnull()] #결측치 제거
```

```python
loan.isna().sum()
```

```python
loan['ISBN'] = loan['ISBN'].astype(str)
loan['ISBN부가기호'] = loan['ISBN부가기호'].astype(str)
```

```python
# 콤마가 있는 값을 제거하는 함수
def remove_comma(value):
    if ',' in value:  # 값에 콤마가 있는지 확인
        return value.replace(',', '')  # 콤마가 있으면 제거
    else:
        return value  # 콤마가 없으면 그대로 반환

# 콤마 제거
loan['ISBN'] = loan['ISBN'].apply(remove_comma)
loan['ISBN부가기호'] = loan['ISBN부가기호'].apply(remove_comma)
```

```python
loan['ISBN'] = loan['ISBN'].astype(float)
loan['ISBN부가기호'] = loan['ISBN부가기호'].astype(float)
```

```python
loan['ISBN'] = loan['ISBN'].astype(int)
loan['ISBN부가기호'] = loan['ISBN부가기호'].astype(int)
loan['KDC'] = loan['KDC'].astype(int)
```
