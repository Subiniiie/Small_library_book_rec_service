```python
from matplotlib.ticker import ScalarFormatter
import numpy as np
x = np.array([2018, 2019, 2020, 2021, 2022])

fig, ax2 = plt.subplots(figsize=(20, 12))

ax2.bar(x,library.iloc[2], alpha=0.5, label='작은 도서관 수', color='#cfcfcf', width=0.8)
ax2.set_ylabel('작은 도서관 수', fontsize=20)
ax2.tick_params(axis='y', labelsize=20)
ax2.tick_params(axis='x', labelsize=20)
ax2.legend(loc='upper left', bbox_to_anchor=(0.00005, 0.97))

ax1 = plt.gca().twinx()
ax1 = plt.gca()
ax1.plot(x, user.iloc[0],color='lightsteelblue', label = '이용자 수',linestyle='solid', linewidth=5, marker='o',markersize=10, markeredgecolor='white', markeredgewidth=3)
ax1.set_xlabel('연도')
ax1.set_title('작은 도서관 수와 이용자 수', fontsize=20)
ax1.set_ylabel('이용자 수', fontsize=20)
ax1.legend(loc='best')
ax1.yaxis.set_major_formatter(ScalarFormatter(useMathText=True))
ax1.ticklabel_format(style='plain', axis='y')
ax1.tick_params(axis='y', labelsize=20)
ax1.tick_params(axis='x', labelsize=20)
ax1.legend(loc='upper left')
# plt.show()

#저장할 때 #plt.show() 하고 해야 함
image_path = '/content/drive/MyDrive/Colab Notebooks/공모전/빅데이터 활용 미래 사회문제 해결 아이디어 공모전/작은 도서관 운영 방안 추천/데이터'
plt.savefig(image_path + '작은 도서관 수와 이용자 수.png', dpi=fig.dpi)
```
