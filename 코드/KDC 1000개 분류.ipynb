```python
INFO_WEIGHT = 0.1

def pearsonR(s1, s2):
    s1_c = s1 - s1.mean()
    s2_c = s2 - s2.mean()

    # NaN 값을 포함하지 않는 유효한 데이터 추출
    valid_data = ~(np.isnan(s1_c) | np.isnan(s2_c))
    s1_c = s1_c[valid_data]
    s2_c = s2_c[valid_data]

    numerator = np.sum(s1_c * s2_c)
    denominator = np.sqrt(np.sum(s1_c ** 2) * np.sum(s2_c ** 2))

    # 0으로 나누는 경우 대응
    if denominator == 0:
        return 0  # 혹은 다른 값을 반환하거나 적절한 조치를 취할 수 있습니다.
    else:
        return numerator / denominator

def recommend(input_book, matrix, n, similar_genre=True):
    input_genres = book[book['서명'] == input_book]['genre'].iloc[0]

    result = []
    for title in matrix.columns:
        if title == input_book:
            continue

        # Rating comparison
        cor = pearsonR(matrix[input_book], matrix[title])

        # genre comparison
        if similar_genre and len(input_genres) > 0:
            temp_genres = book[book['서명'] == title]['genre'].iloc[0]

            same_count = np.sum(np.isin(input_genres, temp_genres))
            cor += (INFO_WEIGHT * same_count)

        if np.isnan(cor):
            continue
        else:
            result.append((title, '{:.2f}'.format(cor), temp_genres))

    result.sort(key=lambda r: r[1], reverse=True)

    return result[:n]
```

```python
recommend_result = recommend('흥칫뿡', matrix, 5, similar_genre=True)
pd.DataFrame(recommend_result, columns=['Title', 'Correlation', 'Genre'])
```
